# Импорт библиотек

```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap
%matplotlib inline
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

 # Получение данных

```python
data = pd.read_csv('../work/input/crimes-in-boston/crime.csv', encoding='utf-8')
data.head()
```

# Проверка на null

```python
data.isnull().sum()
```

# Некоторые максимальные показатели

```python
max_street_crime = data['STREET'].value_counts().index[0]
max_year_crime = data['YEAR'].value_counts().index[0]
max_hour_crime = data['HOUR'].value_counts().index[0]
max_month_crime = data['MONTH'].value_counts().index[0]
max_day_crime = data['DAY_OF_WEEK'].value_counts().index[0]

month = ['Январь','Февраль','Март','Апрель','Май','Июнь','Июль',
         'Август','Сентябрь','Октябрь','Ноябрь','Декабрь']

print('Улица с наиболее высокой частотой преступлений:', max_street_crime)
print('Год, в который произошло наибольшее количество преступлений:', max_year_crime)
print('Час, в который происходит наибольшее количество преступлений:', max_hour_crime)
print('Месяц, в который происходит наибольшее количество преступлений:', month[max_month_crime-1], max_month_crime)
print('День недели, в который происходит наибольшее количество преступлений:', max_day_crime)
```

# Ежегодные показатели преступности

```python
plt.subplots(figsize=(15,6))
sns.countplot('YEAR',data=data,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.title('Количество преступлений каждый год')
plt.show()
```

# Почасовые показатели преступности

```python
plt.subplots(figsize=(15,6))
sns.countplot('HOUR',data=data,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.title('Количество преступлений каждый час')
plt.show()
```

# Распределение тяжких преступлений

```python
plt.subplots(figsize=(15,6))
sns.countplot('OFFENSE_CODE_GROUP',data=data,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.title('Количество тяжких преступлений')
plt.show()
```

# Распределение преступлений по районам города

```python
plt.subplots(figsize=(15,6))
sns.countplot('DISTRICT',data=data,palette='RdYlGn',edgecolor=sns.color_palette('Paired',20),order=data['DISTRICT'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Количество преступлений по районам')
plt.show()
```

# Взаимосвязь между видами преступлений и районами города

```python
pd.crosstab(data.DISTRICT,data.OFFENSE_CODE_GROUP).plot.barh(stacked=True,width=1,color=sns.color_palette('RdYlGn',5))
fig=plt.gcf()
fig.set_size_inches(15,6)
plt.show()
```

# Взаимосвязь между видами преступлений и годом

```python
pd.crosstab(data.YEAR,data.OFFENSE_CODE_GROUP).plot.barh(stacked=True,width=1,color=sns.color_palette('RdYlGn',5))
fig=plt.gcf()
fig.set_size_inches(15,6)
plt.show()
```

# Связь между годом и районом города (количество преступлений)

```python
G1=data[data['MONTH'].isin(data['MONTH'].value_counts()[1:11].index)]
pd.crosstab(G1.YEAR,G1.DISTRICT).plot(color=sns.color_palette('dark',6))
fig=plt.gcf()
fig.set_size_inches(15,6)
plt.show()
```

# Тепловая карта вандализма

```python
vand=data.loc[data.OFFENSE_CODE_GROUP=='Vandalism'][['Lat','Long']]
vand.Lat.fillna(0, inplace = True)
vand.Long.fillna(0, inplace = True) 

BostonMap=folium.Map(location=[42.356145,-71.064083],zoom_start=11)
HeatMap(data=vand, radius=16).add_to(BostonMap)

BostonMap
```
